```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
```

```python
t = np.arange(-3, 3, 0.01)
s = 1 + np.cos(2*np.pi*t)
plt.plot(t, s, '.-g')
plt.plot(t, t*t, '-r')
plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid(True)
```
